**1. Object Count**

-   *Example query:*

    Show count of *‘Classes’* per *‘Amphora form’*, per *'Origin h1'*, per *'Origin h2'*, per *‘Year’* in an amphora
    date range.

-   *Steps:*

    1.  Identify the list of (non unique for now) *‘Classes’* per
        *‘Amphora form’* per *‘Origin’* (h1, h2).

        To that end, *‘Classes’* are grouped by *‘Amphora form’*,
        its *‘Start date’*, *‘And date’* and
        *‘Origins’* (h1, h2). This is done with `pd.groupby()` function. To get
        a list of strings with *‘Class’* names separated by comma,
        `split()` method is used.

        *Example*: For `Amphora form == consp. 20` present
        in `Origin h1 == italy` and `Origin h2 == arezzo`  with date range $-10.0 - 50.0$, the
        list of *‘Classes’* is `[vajilla fina]`.
    
    2.  Define a dictionary with years (as keys), ranging from *‘Amphora
        lower date’* to *‘Amphora upper date’* (across the dataset for a
        given *‘Provenance’*), and zeros (as values).

        Further, for each year the site count will be added as a value.

        *Example:* In Sonata dataset, there are 7 *‘Amphora types’*
        present in *‘Egypt’*. Therefore, 7 amphora date ranges. As a
        result, a dictionary will contain years ranging from $-25$ to
        $499$,

        \begin{equation}
        dict = \{-25:0,  -24: 0, \ldots , 499: 0\}.
        \end{equation}

    2.  Define a dictionary with years (as keys), ranging from *‘Amphora
        start date’* to *‘Amphora end date’* (across the dataset for
        given *‘Origins’*), and zeros (as values).

        Further, for each year the class count will be added as a value.

        *Example:* There are 7 *‘Amphora forms’*
        present in `Origin h1 == italy` and `Origin h2 == arezzo`. Therefore, 7 amphora date ranges. As a
        result, a dictionary will contain years ranging from $-10$ to
        $150$,

        \begin{equation}
        dict = \{-30:0,  -29: 0, \ldots , 150: 0\}.
        \end{equation}

    3.  Loop over years (in a dictionary) and amphora date ranges (in a
        dataframe).

        1.  Check whether a year falls within a date range of each
            *‘Amphora form’*,

            \begin{equation}
            \textit{Amphora start date} <= \textbf{year} <= \textit{Amphora end date.}
            \end{equation}

        2.  If the condition is `False` (it does not fall), then the
            looping process continues (goes to the next amphora date
            range).

        3.  If the condition is `True` (it falls), the list of *‘Classes'*
            for an *‘Amphora form’* is saved in a predefined list. Let’s
            call it *‘result’*.

    4.  Once the iteration over all amphoras’ date ranges for the first
        year in a dictionary is finished, *‘result’* contains from $0$
        to $N$ different lists of *‘Classes’*.

        For instance,

        \begin{equation}
        [['site', 'site'], ['site'], \ldots , ['site', 'site', 'site']].
        \end{equation}

        1.  If the length of result $== \textbf{0}$, then **NO**
            *‘Amphora forms’* were found in a given year. One does not
            need to add $0$ to a dictionary since $0$ value is already
            assigned to each year (see step **2**).

        2.  If the length result $\geq \textbf{1}$, then unique *‘Class’*
            names are identified and counted. For the former, `set()`
            function is employed. For the latter, the length of the list
            with *‘Classes’* is calculated with `len()` method.

        3.  The resulting *‘Class’* count is added to the corresponding
            year in a dictionary as a value.

**2. Summed Frequency**

-   *Example query:*

    Show sum of *‘Frequency’* per *'Origin'* per *‘Site’* per *‘Year’*.

-   *Steps:*

    1.  Group *‘Amphora forms’* by *'Origin'* and *‘Site’*.

        This is done with `pd.groupby()` function.

    2.  Sum *‘Frequencies’* across each *‘Amphora form’* belonging to a certain *'Origin'* and found at
        a particular *‘Site'*.

        To that end, the `sum()` method is used.

    3.  Calculate *‘Frequency’* per *‘Year’* for each *‘Amphora form*
        (with a certain *'Origin'* and *‘Site'*).

        The equation is written as,

        \begin{equation}
        \frac{\text{Amphora form summed frequency}}{\text{Amphora form date range}}.
        \end{equation}

        Here an amphora date range is defined as,

        \begin{equation}
        \textit{Amphora start date} - \textit{Amphora end date}.
        \end{equation}

    4.  Define a dictionary with years (as keys), ranging from *‘Amphora
        start date’* to *‘Amphora end date’* (across the dataset for a
        given *‘Origin’* and *'Site'*), and zeros (as values).

        Further, for each year *‘Amphora form’* summed *‘Frequency’* per
        *‘Year’* will be added as a value.

    5.  Iterate over amphora forms’ date ranges (related to a given
        *‘Origin’* and *''Site*).

        Assign *‘Amphora form’* summed *‘Frequency’* per *‘Year’*
        value to the corresponding year in a dictionary. If several
        *‘Amphora forms’* fall within the same year, their summed
        *‘Frequency’* values per *‘Year’* will be added up.
