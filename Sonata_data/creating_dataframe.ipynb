```python
import pandas as pd
```

#### Loading in dataset and site and contents information 

```python
df = pd.read_csv('SonataDataNewNew.csv')

contents = pd.read_excel('site_contents_data/amphora_content.xlsx', na_filter = False)

site = pd.read_excel(
    'site_contents_data/Template_sites_PK2.xlsx', 
    na_filter = False, 
    sheet_name = 'Sheet1', 
    usecols=['Site', 'NewName', 'Type', 'References and notes']
)
```

```python
print(len(site))
```

```python
df['Site'] = df['Site'].replace({'Romanità di Cattolica ': 'Romanità di Cattolica'})
```

```python
site = site[site.Site != 'Pyrgi']
```

```python
print(len(site))
```

```python
site = site.rename(columns = {
                       'NewName': 'Grouped_sites',
                       'Type':'Site_type', 
                       'Site': 'site',
                       'References and notes':'Site_notes'
                       })

site.head()
```

```python
contents = contents.rename(columns = {
                               'Amphora type': 'amphora_type', 
                               'notes': 'Amphora_type_notes'
                               })
contents.head()
```

### MERGING SITE AND DF

```python
df_merged = pd.merge(df, site, how = 'left', right_on = 'site', left_on = 'Site')
```

```python
df_merged.head()
```

### MERGING CONTENTS TO THE OTHER TWO

```python
df_merged['Amphora_type_main'] = df_merged['Amphora_type'].str.split("_", n = 1, expand = True)[0]
```

```python
df_merged['Amphora_type_main'] = df_merged['Amphora_type_main'].replace({
                                                                    'Dr 2-5 ': 'Dr 2-5', 
                                                                    'Dr 7-11 ': 'Dr 7-11', 
                                                                    'Garloise 1': 'Gauloise 1', 
                                                                    'Dr 20 ':'Dr 20', 
                                                                    'Ostia LIX/XXIII': 'Ostia XXIII/LIX'
                                                                    })
```

```python
df_merged1 = pd.merge(df_merged, contents, how = 'left', right_on = 'amphora_type', left_on = 'Amphora_type_main')
```

```python
df_merged1.head()
```

```python
df_merged1.to_csv('Sonata_data_site_contents.csv')
```
